```python
### Import Necessary Libraries
import requests
```

```python
### User Inputs
summoner_region = 'NA1'
summoner_name = 'Simply Dadrian'
```

```python
### Revolving 24-hour API Key: https://developer.riotgames.com/
api_key = ''

### API URL for ACCOUNT-V1 to get puuid from `summoner_name` and `summoner_region`
api_url_account = f'https://americas.api.riotgames.com/riot/account/v1/accounts/by-riot-id/{summoner_name}/{summoner_region}?api_key={api_key}'
player_uuid = requests.get(api_url_account).json()['puuid']

### API URL for TFT-MATCH-V1 to get match history for player
api_url_matches = f'https://americas.api.riotgames.com/tft/match/v1/matches/by-puuid/{player_uuid}/ids?api_key={api_key}'
matches = requests.get(api_url_matches).json()
match_id = matches[0]

### API URL for TFT-MATCH-V1 to get granular match data for player
api_url_match = f'https://americas.api.riotgames.com/tft/match/v1/matches/{match_id}?api_key={api_key}'

### Validate API Connections
print(f'ACCOUNT-V1 API Status: {requests.get(api_url_account)}')
print(f'TFT-MATCH-V1 Match History API Status: {requests.get(api_url_matches)}')
print(f'TFT-MATCH-V1 Match Granular API Status: {requests.get(api_url_match)}')
print(f'\nPlayer UUID: {player_uuid}')
print(f'Previous 20 Matches: {matches}')
print(f'Most Recent Match: {match_id}')
```

```python
### Getting JSON Data
match_data = requests.get(api_url_match).json()
part_index = match_data['metadata']['participants'].index(player_uuid)
placement = match_data['info']['participants'][part_index]['placement']

print(f'Placement: {placement}')
```

```python
def get_match_data(region):
    api_url = f'https://{region}.api.riotgames.com/tft/match/v1/matches/{match_id}?api_key={api_key}'
        
    resp = requests.get(api_url)
    data = resp.json()
    return data
```
